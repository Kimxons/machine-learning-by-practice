```python
# Data visualization

# You must understand your data in order to get the best results from
# machine learning algorithms.

# The fastest way to learn more about your data is to use data
# visualization.

# Univariate plots

# Three techniques that one can use to understand each
# attribute of a dataset independently include:
#  Histograms.
#  Density Plots.
#  Box and Whisker Plots.

# Boxplots summarize the distribution of each attribute, drawing a line
# for the median (middle value) and a box around the 25th and 75th
# percentiles (the middle 50% of the data).

# The whiskers give an idea of the spread of the data and dots outside
# of the whiskers show candidate outlier values (values that are 1.5
# times greater than the size of spread of the middle 50% of the data).
```

```python
# Univariate Histograms
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("pima/diabetes.csv")
# data.head()
# data.columns

data.hist()
plt.show()
```

```python
# Univariate Density plots
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.show()
```

```python
#Univariate Box and Whisker Plots
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```

```python
# Multivariate plots show the interactions between multiple variables in your dataset.
#  *Correlation Matrix Plot.
#  *Scatter Plot Matrix.

# a. Correlation Matrix Plot.
# Correlation gives an indication of how related the changes are between
# two variables.
```

```python
#multivariate correlation matrix plot
correlations = data.corr()
# print(correlations)

# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1,vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns)
ax.set_yticklabels(data.columns)
plt.show()
```
